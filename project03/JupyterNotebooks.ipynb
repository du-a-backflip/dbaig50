## What is Jupyter Notebook?
<b>Jupyter Notebook</b> is an open-source web app that allows the user to create, edit, and share computational notebooks
* Computational notebooks are sharable and interactive documents that can contain computer code, data, text, figures, charts, etc. all in one place!

It is part of <b>Project Jupyter</b>, which is a project committed to developing open-source software, open-standards, and tools for interactive computing for multiple programming languages. It was founded by Fernando Pérez and Brian Granger in 2014.
* <b>Open-standards</b> are protocols that are accessible to the public. Examples include:
    * 4G and 5G
    * GSM (Global System for Mobile Communications)

Initially, Jupyter Notebook was called iPython Notebook. However, as more languages were added and the Jupyter Project came into being, it was renamed to Jupyter Notebook.

## Cells
In Jupyter Notebook, the user types into blocks (multi-line input text fields) called cells.

There are three kinds of cells:
1. <b>Code Cells</b> - where the user can type lines of code.
    * They are marked with square brackets [] beside them. When run, their output is shown beneath (or beside) them
2. <b>Markdown Cells</b> - where the user can input formatted text. 
3. <b>Raw Cells</b> - where the user can input unformatted text.
    * These are only really useful if the user wants to use nbconvert (a command line tool) to convert the notebook to another format (eg: LaTeX or HTML)

At one point there were also heading cells. However, they have since been phased out, since the user can just make headings in the Markdown Cells. (RIP)

By default, all cells are code cells. However, the user can change that by clicking where it says “Code” in the toolbar. Clicking there  opens a dropdown where users can select the type of cell they wish to use.


## Kernels
A kernel is a computational engine that executes whatever code written in the notebook. By default, Jupyter Notebook comes with a kernel for Python 3 (known as ipykernel). Kernels for other languages need to be manually installed.

If the user wants, multiple notebooks can use the same kernel. This can come in handy if your notebook becomes too long and you want to split it, or if you have multiple notebooks using the same data structures and variables. 

The Jupyter kernel is made up of: the kernel process, the kernel manager, and the kernel gateway
1. The <b>kernel process</b> is what executes the code in the notebook. It also returns the results of said code to displayed on the frontend for the user.
2. The <b>kernel manager</b> starts, stops, and restarts the kernel when needed.
3. The <b>kernel gateway</b> connects the kernel to the front end of the notebook over a network connection

The kernel also uses <b>WebSockets</b> to help the frontend and kernel communicate with each other. The WebSocket connection is established when you open the notebook and start the new kernel.




## Benefits
Some benefits of Jupyter Notebooks include:
* The ability to split lengthy code into cells
     * This allows the user to test small aspects of their code instead of having to run a whole script over and over again.
* The output of code cells being shown directly under (or next to) the cell responsible for them
     * This allows the user to easily see what each code cell is doing, and makes it easier to experiment with code.
* The ability to save output of code cells
     * This makes it so that the user doesn't have to run a script whenever re-opening the notebook
* Aid in communication/presentation of code (as explanatory text and output can be provided with code)
* It having widgets (interactive elements) that can be used for data science.

Because of these benefits, Jupyter Notebook is typically used by data scientists and ML practitioners. It is also used by students as a tool to learn code with.
